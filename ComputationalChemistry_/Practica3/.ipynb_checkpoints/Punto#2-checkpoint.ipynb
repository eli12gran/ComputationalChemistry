En este item se nos pide realizar un análisis de la estructura de bandas y la densidad de estados (DOS) del material. Para esto se hará uso de la imagen que aparece en Materials Project al desplegar la información dell material. Esta es:


```python
import pymatgen
import pandas as pd
from pymatgen.core import Structure, Lattice
import crystal_toolkit
from pymatgen.ext.matproj import MPRester
from pymatgen.transformations.standard_transformations import AutoOxiStateDecorationTransformation
from pymatgen.electronic_structure.plotter import DosPlotter
from pymatgen.electronic_structure.plotter import BSPlotter
```

```python
api_key = 'YVb1EJfpW6q8GgufwhezGRB7YsGcbEXS'
mpr = MPRester(api_key)
dos = mpr.get_dos_by_material_id('mp-4571')

# Grafica DOS
plotter = DosPlotter()
plotter.add_dos('Total DOS', dos)
plotter.show(xlim=[-5, 5])
```

```python
band_structure = mpr.get_bandstructure_by_material_id('mp-4175')
bs_plotter = BSPlotter(band_structure)
bs_plotter.get_plot(vbm_cbm_marker=True)
```
