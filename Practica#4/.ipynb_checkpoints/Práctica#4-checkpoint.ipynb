### **Item b - Item c**

Para la creación del diccionario con las energías libres de superficie lo que hacemos es escribir la superficie con los indices de Miller (h,k,l) en forma de tupla y luego a estos les asignamos, usando los dos puntos como se debe realizar en los diccionarios, el valor de las energías dadas en la hoja de la práctica.

Debajo del diccionario crearemos la estructura primitiva del material utilizando la función `bulk()` de ASE. A esta función le debemos pasar el nombre del metal al cual le queremos construir la celda primitiva (en este caso **Fe**), luego su estructura cristalógrafica, (en este caso **fcc**), luego el valor del parámetro **a**, (en este caso $2.86$) y por último un valor booleano que confirma que la estructura en este caso es cúbica. Despues de la siguiente línea de códgio se explicará cómo se llegó a cada una de estas conclusiones sobre la simetría del metal.

Esto es:

```python
from ase.build import bulk
Fe_surf = {(1,1,0): 1.371 , (2,0,0): 1.397 , (2,1,1): 1.608 }
bulk_Fe = bulk("Fe", crystalstructure= "fcc", a=2.86, cubic=True)
view(bulk_Fe)
```

Usando la información que provee Materials Project, podemos observar el valor del parámetro `a`; dado que el sistema es **cúbico** los parámetros $a,b,c$ tienen exactamente el mismo valor, $2.86$, al igual que sus ángulos, que son todos $90°$. Para saber qué tipo de simetría/dónde está centrado el sistema, podemos utilizar su grupo espacial que se puede obtener usando `pymatgen` de materials project. Esto es:



```python
Fe_materials.get_space_group_info()
```

Utilizando la información de **space group** de la página de wikipedia, podemos ver que este grupo se encuentra entre las **simetrías cúbicas centradas en la cara**. 



### **Item d**

Para crear las dos superficies de Wulff utilizaremos el paquete `wulffpack` cargado en la primera línea de código de la práctica. Para esto utilizaremos la función `SingleCrystal` que recibe varios parámetros; como la mayoría de las funciones en python, algunos parámetros no son obligatorios, por lo que solo incluiremos algunos. Estos serán:

* Energías de superficie: `surface_energies = Fe_surf`

* Estructura Primitiva: `primitive_structure = bulk_Fe`

* Número de átomos: `natoms = 3000`

Anteriormente habíamos definido nuestra estructura primitiva y el diccionario de energías de superficie, por lo que directamente podemos asignar estas variables a la función. Esto es:

```python
import wulffpack 
from wulffpack import SingleCrystal
surface_1 = SingleCrystal(Fe_surf, bulk_Fe, natoms = 3000)
surface_1.view()
```

Para la superficie dos tendremos los mismos parámetros que mencionamos anteriormente solo que el **`natoms`** cambiará. Ya su valor no será de $3000$ sino de $1000$. Esto es:

```python
surface_2 = SingleCrystal(Fe_surf, bulk_Fe, natoms = 10000)
surface_2.view()
```

### **Item d**

La superficie que escogeremos para este punto será la de mayor intensidad, es decir, la que tiene índices de Miller **(1,1,2)**. A partir de esta superfice crearemos las dos superceldas. Para esto lo que haremos será tomar la superficie ya creada en la variable `CaZrO3_surf_1_1_2` que corresponde a un **Slab 1x1** con $3$ capas y $10A$ de distancia y la multiplicaremos por una tupla que tendrá los valores correspondientes a $(x,y,z)$. En estos casos el eje $z$ no se cambia, lo dejamos sin "repetir" mientras que el eje $x$ y $y$ cambiarán por los números para la primera superficie por $2$ y $3$ y para la segunda superficie por $3$ y $4$. 

Veamos esto cómo nos queda:

```python
CaZrO3_2x3 = CaZrO3_surf_1_1_2 * (2,3,1)
view(CaZrO3_2x3)
```

```python
CaZrO3_3x4 = CaZrO3_surf_1_1_2 * (3,4,1)
view(CaZrO3_3x4)
```
