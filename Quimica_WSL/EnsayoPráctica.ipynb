```python
from ase.io import write, read
from ase.visualize import view
from ase.build import bulk
from ase.build import surface
```

```python
# Cu2O
Cu2O = read("Cu2O.cif")
view(Cu2O)
```

```python
Cu2O_1_2_1 = surface(Cu2O, (1,2,1), 4, 10)
view(Cu2O_1_2_1)
```

```python
supercell = Cu2O_1_2_1 * (2,2,1)
view(supercell)
```

```python
write("Cu2O_supercell.cif", supercell)
```
