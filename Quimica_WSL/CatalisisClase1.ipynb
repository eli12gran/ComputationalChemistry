```python
from ase.io import write, read
from ase.visualize import view
from ase.build import bulk
from ase.build import surface
```

```python
Pt = read("Pt.cif")
view(Pt)
```

```python
Pt_1_1_1 = surface(Pt, (1,1,1), 4, 10)
view(Pt_1_1_1)
```

```python
Pt_2x2 = Pt_1_1_1 * (2,2,1)
view(Pt_2x2)
```

```python
write("Pt_2x2.cif", Pt_2x2)
```
